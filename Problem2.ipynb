```python

# Triangle class 

class triangle:

    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c

    def validate_triangle(self,a,b,c):
        if ((self.a**2)+(self.b**2)) == (self.c**2):
            print ('valid triangle')

        else:
            print('invalid triangle')


# Rectangle class

class rectangle:

    def __init__(self,a,b,c,d):
        self.a = a
        self.b = b
        self.c = c
        self.d = d

    def validate_rectangle(self,a,b,c,d):
        if (a == c) & (b == d):
            print ('valid rectangle')

        else:
            print('invalid rectangle')


# Creating triangle object:

Triangle = triangle(3,4,5)

# Creating rectangle object:

Rectangle = rectangle(2,4,2,4)


# Validating rectangle & triangle object

Triangle.validate_triangle(3,4,5)
Rectangle.validate_rectangle(2,4,2,4)
```
